```python
# Get column names as a list
column_names = data.columns.tolist()

# Print column names
print("Column names:", column_names)
```

```python
import pandas as pd

# Create a Pandas Excel writer using XlsxWriter as the engine.
with pd.ExcelWriter('Data_Summary.xlsx', engine='xlsxwriter') as writer:
    # Write each DataFrame to a different worksheet.
    Describe.to_excel(writer, sheet_name='Describe')
    types.to_frame(name='Data Types').to_excel(writer, sheet_name='Data Types')
    tails.to_excel(writer, sheet_name='Tail')
    heads.to_excel(writer, sheet_name='Head')
    unique_counts.to_frame(name='Unique Counts').to_excel(writer, sheet_name='Unique Counts')

    # Use the writer to save a DataFrame of column names to a named sheet.
    pd.DataFrame({'Column Names': column_names}).to_excel(writer, sheet_name='Column Names', index=False)
```

```python
# Example to convert a column to numerical type
for i in column_names:
  data[i] = pd.to_numeric(data[i], errors='coerce')
  print("done" + i)
```
